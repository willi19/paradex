```python
import os
import subprocess
from paradex.utils.path import shared_dir, home_path
from paradex.utils.upload_file import rsync_copy

vid_name = ["23280285.avi"]
demo_root_path = os.path.join(shared_dir, "capture/miyungpa")

failed = []

for obj_name in os.listdir(demo_root_path):
    index_list = os.listdir(os.path.join(demo_root_path, obj_name))
    for index in index_list:
        demo_path = os.path.join("capture/miyungpa", obj_name, index)
        print(f"Processing {demo_path}...")
        
        video_path = os.path.join(shared_dir, demo_path, "videos", vid_name[0])
        save_path = os.path.join(home_path, "paradex_download", demo_path, vid_name[0])
        
        # Copy video
        rsync_copy(video_path, save_path)
        
        # Extract frames
        vid_basename = os.path.splitext(vid_name[0])[0]  # "23280285"
        output_dir = os.path.join(home_path, "paradex_download", demo_path, "images", vid_basename)
        os.makedirs(output_dir, exist_ok=True)
        
        output_pattern = os.path.join(output_dir, "%05d.jpg")
        
        cmd = [
            "ffmpeg",
            "-i", save_path,
            "-q:v", "2",  # High quality
            output_pattern
        ]
        
        print(f"  Extracting frames to {output_dir}...")
        try:
            subprocess.run(cmd, check=True)
        except subprocess.CalledProcessError as e:
            print(f"  Error extracting frames: {e}")
            failed.append(demo_path)
            continue
        print(f"  âœ“ Done")
print("Failed cases:")
for case in failed:
    print(case)
```
